#### Q1. What is Flask Framework? What are the advantages of Flask Framework?
Flask is a micro web framework written in Python. It is classified as a microframework because it does not require particular tools or libraries. It has no database abstraction layer, form validation, or any other components where pre-existing third-party libraries provide common functions. However, Flask supports extensions that can add application features as if they were implemented in Flask itself. Flask is a lightweight WSGI web application framework. It is designed to make getting started quick and easy, with the ability to scale up to complex applications. It began as a simple wrapper around Werkzeug and Jinja and has become one of the most popular Python web application frameworks.

Advantages of Flask Framework:
1. `Easy to use:` Flask is easy to use and lightweight. It does not require any specific tools or libraries.
2. `Flexibility:` Flask is flexible and allows developers to use the tools and libraries they want.
3. `Scalability:` Flask is scalable and can be used to build simple or complex applications.
4. `Extensible:` Flask is extensible and allows developers to add features using extensions.
5. `Active community:` Flask has an active community that provides support and resources for developers.
6. `Built-in development server:` Flask comes with a built-in development server that makes it easy to test applications.

#### Q3. What is App routing in Flask? Why do we use app routes?
App routing in Flask refers to the process of mapping URLs to view functions. It allows developers to define routes that specify the URL patterns and the corresponding view functions that should be executed when a request is made to a specific URL. App routing is an essential feature of Flask that enables developers to create web applications with multiple pages and functionalities.
We use app routes for several reasons:
1. `URL mapping:` App routes allow developers to map URLs to view functions, making it easy to define the structure of the web application.
2. `Request handling:` App routes specify the view functions that should be executed when a request is made to a specific URL, allowing developers to handle requests and generate responses.
3. `Page navigation:` App routes enable developers to create multiple pages and navigate between them by defining different routes for each page.
4. `Dynamic content:` App routes can be used to pass dynamic data to view functions, allowing developers to generate dynamic content based on user input or other factors.
5. `Error handling:` App routes can be used to define error handlers that handle exceptions and errors that occur during the execution of the web application.

#### Q4. Create a `/welcome` route to display the welcome message `Welcome to ABC Corporation` and a `/` route to show the following details: 
```python
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999
```

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the `url_for()` function.
The `url_for()` function is used in Flask for URL building. It generates a URL for the specified endpoint by using the endpoint name and any arguments provided. The `url_for()` function takes the endpoint name as the first argument and any additional arguments as keyword arguments. It returns the URL for the specified endpoint.
Here is a Python code to demonstrate the working of the `url_for()` function:

````python
from flask import Flask, url_for
app = Flask(__name__)


@app.route('/')
def index():
    return 'Index Page'


@app.route('/login')
def login():
    return 'Login Page'


@app.route('/user/<username>')
def profile(username):
    return f'{username}\'s profile'


if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('index'))
        print(url_for('login'))
        print(url_for('login', next='/'))
        print(url_for('profile', username='John Doe'))

    app.run()
````

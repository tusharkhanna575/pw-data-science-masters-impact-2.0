### Q.1  Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1. **Scatter Plot**: Used to visualize the relationship between two continuous variables. It helps in identifying correlations and patterns.
2. **Line Plot**: Used to display trends over time or ordered categories. It connects data points with a line, making it useful for time series data.
3. **Bar Plot**: Used to compare categorical data with rectangular bars. It shows the frequency or value of each category, making it easy to compare different groups.
4. **Box Plot**: Used to summarize the distribution of a dataset through its quartiles. It highlights the median, quartiles, and potential outliers, providing insights into data spread and skewness.
5. **Heatmap**: Used to visualize data through color coding in a matrix format. It is particularly useful for displaying correlation matrices or frequency tables, allowing for quick identification of patterns and relationships.

###  Q.2 Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using `x = "timepoint"` and `y = "signal"` for different events and regions. 

Note:  timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the fmri dataset
fmri = sns.load_dataset("fmri")
# Create a line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("FMRI Signal Over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event and Region")
plt.show()
```

### Q.3  Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using `x = 'pclass'`, `y = 'age'` and `y = 'fare'`. 

Note:  pclass, age, and fare are columns in the titanic dataset.

```python
# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Create box plots for 'age' and 'fare' by 'pclass'
plt.figure(figsize=(12, 6))
sns.boxplot(data=titanic, x="pclass", y="age")
plt.title("Box Plot of Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
plt.figure(figsize=(12, 6))
sns.boxplot(data=titanic, x="pclass", y="fare")
plt.title("Box Plot of Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

#### Q.4  Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")
# Create a histogram for the 'price' column with hue for 'cut'
plt.figure(figsize=(12, 6))
sns.histplot(data=diamonds, x="price", hue="cut", multiple="stack", bins=30)
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Count")
# plt.legend(title="Cut")
plt.show()
```

####  Q.5 Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.  

```python
import seaborn as sns
# Load the iris dataset
iris = sns.load_dataset("iris")
# Create a pair plot with hue for species
sns.pairplot(iris, hue="species")
plt.suptitle("Pair Plot of Iris Dataset by Species")
plt.show()
```

#### Q.6 Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
# Load the flights dataset
flights = sns.load_dataset("flights")
# Pivot the dataset to create a matrix for the heatmap
flights_pivot = flights.pivot(
    index="month", columns="year", values="passengers")
# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Heatmap of Passengers by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```

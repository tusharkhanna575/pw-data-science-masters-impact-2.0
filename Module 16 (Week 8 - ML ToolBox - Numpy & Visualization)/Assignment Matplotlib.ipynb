##### Q.1 What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK.

Matplotlib is used for creating static, animated, and interactive visualizations in Python. It can be used in Python scripts, the Python and IPython shell, web application servers, and various graphical user interface toolkits.

Five types of plots that can be created using the Pyplot module of Matplotlib are:

1. Line plot
2. Bar plot
3. Histogram
4. Scatter plot
5. Pie chart

#### Q.2 What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```
Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data. The data are displayed as a collection of points, each having the value of one variable determining the position on the horizontal axis and the value of the other variable determining the position on the vertical axis. This kind of plot is also called a scatter chart, scattergram, scatter diagram, or scatter graph. Scatter plots are used to observe relationships between variables.

````python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('title')
plt.scatter(x, y)
plt.show()
````

#### Q.3 Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

```python   
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

The subplot() function is used to create multiple plots in the same figure. It is a convenient method to create multiple plots in a single figure. The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the subplot.

````python
import matplotlib.pyplot as plt
import numpy as np

# Data for the lines
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
fig, axs = plt.subplots(2, 2)

# Plot each line in a different subplot
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x, y2, 'tab:orange')
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x, y3, 'tab:green')
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x, y4, 'tab:red')
axs[1, 1].set_title('Line 4')

# Add a title for the whole figure
fig.suptitle('Four line plots')

# Display the plot
plt.show()
````

#### Q.4 What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

A bar plot is a plot that presents categorical data with rectangular bars with lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally. 
A bar plot is used to compare data across different categories. It is used to visualize the distribution of a categorical variable.

````python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.title("Vertical bar chart")
plt.show()

# horizontal bar chart
plt.barh(company, profit)
plt.title("Horizontal bar chart")
plt.show()
````

#### Q.5 What is a box plot? Why is it used? Using the following data plot a box plot.
```python
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

A box plot is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile, median, third quartile, and maximum. 
It is used to identify the central tendency, variability, and skewness of the data distribution. Box plots are also known as box-and-whisker plots.

````python
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1, box2])
plt.title("Boxplot")
plt.show()
````
